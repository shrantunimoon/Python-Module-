```python
#error handling: helps us create code that identifies future errors and give a helpful
response  to the user ,so that they can work better with the code
#TypeError
#error handling is done using try except block
try:
    a=int(input('provide value'))
    for i in range(a):
        print(i) 
except:
    print('give integer value')
```

```python
#nested try except blocks
try:
    a='manish'
    for i in range(len(a)):
        print(i) 
    try:
        b=input('provide 1st value')
        b=int(b)
        m=input('provide 2nd value')
        m=int(m)
        print(b/m)
    except ZeroDivisionError:
        print('give proper 2nd  value')
except (TypeError,ValueError):
    print('give integer value')  
except:
    print('check input')
```
