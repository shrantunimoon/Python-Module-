```python
#hiding : encapsulation : hiding variable
class A:
    def __init__(self,a,b):
        self.a=a
        self.__b=b #hidding the variable using double underscore
    def getter(self): #to get the valueof the hidden variable
        print(self.__b) 
    def setter(self,new): #to set a new value of the hidden method
        self.__b=new
        
m=A(12,13) 
m.a
#m.__b #name mangling 
m.getter()
m.setter(90)
m.getter()
```

```python
from abc import abstractmethod,ABC
class A(ABC):
    @abstractmethod
    def fun1(self):
        print('this is fun1')
    def fun2(self):
        print('g') 
        
class subclass(A):
    def fun1(self):
        print('this is subclass') 

#m=A()
m=subclass()
m.fun1()
```

```python
from abc import ABC,abstractmethod
class ATM(ABC):
    @abstractmethod
    def pinchange(self):
        pass
   
    def withdraw(self):
        pass
    def check(self):
        pass

class subclass(ATM):
    def pinchange(self):
        print('this is pinchange')  
    def withdraw(self):
        print('this is witdraw')
    def check(self):
        print('check')


        

f=subclass()
f.check()
```
