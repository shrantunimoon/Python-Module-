```python
def disc(s):
    if s>450 and s<699:
        return s*0.1
    elif s>700 and s<1199:
        return s*0.2
    elif s>1200:
        return s*0.3
    else:
        return s
        


m=eval(input('give your list'))
list(map(disc,m))
```

```python
from random import randint, choice
def passwordgen():
    password = ''
    specialcharacters = list("!@#$%^&*?()")
    alpha = list("abcdefghijklmnopqrstuvwxyz")
    i = 0
    while i<10:
        password += choice((choice(specialcharacters),choice(alpha),str(randint(0,9))))
        i+=1
    print(password) 

passwordgen()
```

```python
def fac(num):
    mul=1
    for i in range(1,num+1):
        mul*=i
    return mul

fac(5)
```

```python
def fact(num):
    if num==1 or num==0:
        return 1
    else:
        return num*fact(num-1)

fact(5)
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```
