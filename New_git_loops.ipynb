## Why do we need loops
Loops let us repeat a task without writing the same code again.

They help process collections of data like lists, files, or records.

They make code shorter, cleaner, and scalable.

#### Difference between for and while in Python

for loop is used when the number of iterations is known (or when iterating over a sequence).

while loop is used when the number of iterations is unknown and depends on a condition.

for automatically handles iteration; while requires manual condition updates.

while can easily lead to infinite loops if the condition is not updated properly.

for is more readable for collections, while is better for condition-based looping
